```python
"""
    Задание 1
    Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.
    Определите какому фильму было выставлено больше всего оценок 5.0.
"""
```

```python
import pandas as pd
```

```python
# необходимые для задачи данные из датасета
data = pd.read_csv('./ml-20m/ratings.csv', usecols=['movieId','rating'])
data = data[ data['rating'] == 5.0 ]
data.head()
```

```python
# количество записей для каждого фильма с рейтингом 5.0
data = data.groupby('movieId', as_index=False).count()
data.head()
```

```python
# вдруг у нескольких фильмов оказался одинаковый рейтинг?
data = data[ data['rating'] == data['rating'].max()]
data.head()
```

```python
# наименования фильмов для удобства
movies = pd.read_csv('./ml-20m/movies.csv', usecols=['movieId', 'title'], index_col=['movieId'])
result = data.merge(movies, on='movieId', how='left')
result.get(['title'])
```

```python
"""
    Задание 2
    По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
    категорий 4, 12 и 21 за период с 2005 по 2010 года.
    Не учитывайте в расчетах отрицательные значения quantity.
"""
```

```python
import pandas as pd
```

```python
countries = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]

data = pd.read_csv('power.csv')
data = data[ (data.country.isin(countries)) & (data.category.isin(categories)) & (data.quantity > 0)]
data[ (data.year >= 2005) & (data.year <= 2010)]
data.quantity.sum()
```
