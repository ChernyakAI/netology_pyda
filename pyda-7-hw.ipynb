```python
"""
    Задание 1
    Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

    Результат классификации запишите в столбец class
"""
```

```python
import pandas as pd
```

```python
def classification(rating):
    if rating <= 2:
        return 'low'
    elif rating <= 4:
        return 'medium'
    else:
         return 'high' 
```

```python
ratings = pd.read_csv('./ml-latest-small/ratings.csv')
ratings['class'] = ratings.rating.apply(classification)
ratings.head()
```

```python
"""
    Задание 2
    Используем файл keywords.csv.

    Необходимо написать гео-классификатор,
    который каждой строке сможет выставить географическую принадлежность определенному региону.
    Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона.
    Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

    Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
    geo_data = {
                'Центр': ['москва', 'тула', 'ярославль'],
                'Северо-Запад': ['петербург', 'псков', 'мурманск'],
                'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }

    Результат классификации запишите в отдельный столбец region.
"""
```

```python
import pandas as pd
```

```python
def geo_classificator(keywords):
    
    for region in geo_data:
        for town in geo_data[region]:
            if town in keywords:
                return region

    return 'undefined'
```

```python
keywords = pd.read_csv('keywords.csv')
geo_data = {
                'Центр': ['москва', 'тула', 'ярославль'],
                'Северо-Запад': ['петербург', 'псков', 'мурманск'],
                'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords['region'] = keywords.keyword.apply(geo_classificator)
# контроль:
keywords[ keywords['region'] != 'undefined']
```

```python
"""
    Задание 3
    Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”.
    Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия.
    Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

    При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:
    1. В переменную years запишите список из всех годов с 1950 по 2010.
    2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска.
        Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
            a) для каждой строки пройдите по всем годам списка years
            б) если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
            в) если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
    3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
    4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга
"""
```

```python
import pandas as pd
```

```python
def production_year(title):   
    for year in years:
        if str(year) in title:
            return year
    return 1900
```

```python
years = [year for year in range(1950, 2011)]

ratings = pd.read_csv('./ml-latest-small/ratings.csv', usecols=['movieId', 'rating'])
movies = pd.read_csv('./ml-latest-small/movies.csv', usecols=['movieId', 'title'], index_col=['movieId'])
ratings = ratings.merge(movies, on='movieId', how='left')
ratings['year'] = ratings.title.apply(production_year)
ratings = ratings.groupby(['year']).mean().sort_values(by='rating', ascending=False).reset_index()
ratings.drop(['movieId'], axis=1, inplace=True)
ratings = ratings[ratings.year != 1900]
ratings
```
