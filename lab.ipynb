Дана строка 
string_01 = “Mama, just killed a man, put a gun against his head, pulled my trigger, now he's dead. Mama, life had just begun, but now I've gone and thrown it all away. Mama, ooh (any way the wind blows), didn't mean to make you cry. If I'm not back again this time tomorrow carry on, carry on as if nothing really matters.” 1.Посчитайте, сколько в строке символов (исключая пробелы).
2.Посчитайте, сколько в строке слов.
3.Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки, заканчивающиеся на эту букву (например, “s”).
4.Не обязательно: Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки, начинающиеся на эту букву (например, “m”).

```python
string_01 = "Mama, just killed a man, put a gun against his head, pulled my trigger, now he's dead. Mama, life had just begun, but now I've gone and thrown it all away. Mama, ooh (any way the wind blows), didn't mean to make you cry. If I'm not back again this time tomorrow carry on, carry on as if nothing really matters."

# 1.Посчитайте, сколько в строке символов (исключая пробелы).
print(f'1. Символов строке, исключая пробелы: {len(string_01) - string_01.count(" ")}')

# 2.Посчитайте, сколько в строке слов.
words_count = len(string_01.split(' '))
print(f'2. Количество слов в строке: {words_count}')

# 3.Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки,
#      заканчивающиеся на эту букву (например, “s”).
def words_end_with_a_letter(letter):
    words = string_01.split(' ')
    result = [word for word in words if word[-1] == letter]
    return result

print(f'3. Слова, заканчивающиеся на букву s: {words_end_with_a_letter("s")}')
      

# 4.Не обязательно: Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки,
#      начинающиеся на эту букву (например, “m”).
def words_start_with_a_letter(letter):
    words = string_01.split(' ')
    result = [word for word in words if word[0] == letter]
    return result

print(f'4. Слова, начинающиеся на букву m: {words_start_with_a_letter("m")}')
```

# Задача 2

Напишите код, который будет генерировать случайную последовательность из 6 цифр, одна из которых должна быть 3. Позиция цифры 3 должна определяться случайным образом. Результат должен быть строкой. Например: "456309" или "330127".

```python
import random

result = ''
for i in range(6):
    letter = random.randint(0, 9)
    result += str(letter)
    if (i == 4) and ('3' not in result):
        result += '3'
        break

print(result)
```

```python
# 2. Напишите функцию, которая принимает в качестве аргумента номер студента и новый номер группы
# и позволяет изменить номер группы этого студента на новый.
def update_data(student_number, group_number):
    for index, data in enumerate(students):
        if data[0] == student_number:
            data[5] = group_number

update_data('1102', 'ffffff')
students
```
